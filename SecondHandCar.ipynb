```python
tensor_df = tf.constant(data, dtype = tf.float32)
tensor_df
```

```python
tensor_df = tf.random.shuffle(tensor_df)
tensor_df
```

```python
X = tensor_df[:, 2:-1]
X
```

```python
y = tensor_df[:, -1]
y = tf.expand_dims(y, axis = -1)
y
```

```python
from tensorflow.keras.layers import Normalization
```

```python
normalizer = Normalization()
normalizer.adapt(X)
normalizer(X)
```

```python
model = tf.keras.Sequential([normalizer, tf.keras.layers.Dense(1)])
model.summary()
```
